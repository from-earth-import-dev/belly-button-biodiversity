```python
app = Flask(__name__)
```

```python
app.config["SQLALCHEMY_DATABASE_URI"] = "sqlite:///db/bellybutton.sqlite"
```

```python
db = SQLAlchemy(app)
```

```python
Base = automap_base()
```

```python
Base.prepare(db.engine, reflect=True)
```
